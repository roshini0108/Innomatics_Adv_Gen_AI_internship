Problem 2: Highest Salary from Employee Data

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
maxe=max(employees,key=employees.get) #gets key with max key value
print(maxe ,'-',employees[maxe])
```

Problem 4: Count Products Above a Price Threshold

```python
prices = [450, 1200, 899, 1500, 300]
count=0
for i in prices:
  if i>1000: #prints no of prices greater then 1000 by counting each elemet greater then 1000
    count+=1
print("Products above 1000:",count)
```

Problem 5: Calculate Attendance Percentage

```python
attendance = ["P", "P", "A", "P", "P"]
count_p=0
total=len(attendance)
for i in attendance:
  if i=='P': #if present increase the present count
    count_p+=1
print("Attendance Percentage:",((count_p/total)*100)) #no.of preset/total no of days *100 gives attendance percentage
```
