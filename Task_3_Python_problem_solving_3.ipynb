Problem Statement 1: Employee Performance Bonus Eligibility


```python
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}
maxe=max(employees.values())
top=[]
for name,score in employees.items():
  if score==maxe:
    top.append(name)
print(f"Top Performers Eligible for Bonus: {', '.join(top)} (Score: {maxe})")
```

Problem Statement 2: Search Query Keyword Analysis


```python
t=input()
t=t.lower()
w=t.split()
freq={}
for wo in w:
  freq[wo]=freq.get(wo,0)+1
filtered_data = {word: count for word, count in freq.items() if count > 1}
print(filtered_data)
```

Problem Statement 3: Sensor Data Validation

```python
liste=list(map(int,input().split(',')))
listee=[]
for index,value in enumerate(liste):
    if value%2==0:
        listee.append((index,value))
print("Valid Sensor Readings (Hour, Value):",listee)
```

Problem Statement 4: Email Domain Usage Analysis


```python
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]
domain_count={}
leng=len(emails)
for email in emails:
    email=email.strip().lower()
    domain=email.split('@')[1]
    domain_count[domain]=domain_count.get(domain,0)+1
for domain, count in domain_count.items():
    percentage = (count / leng) * 100
    print(f"{domain}: {percentage:.2f}%")
```

Problem Statement 5: Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
average = sum(sales) / len(sales)
threshold = average * 1.25
for index, value in enumerate(sales):
    if value >= threshold:
        print(f"Day {index + 1}: {value}")
```
