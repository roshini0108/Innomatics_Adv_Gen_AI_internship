Problem 1: Social Media – Post Engagement Analyzer

Problem 3: Agriculture – Rainfall Adequacy Checker

```python
def seasonal_rainfall_data(average_rainfall,required_rainfall):
  if average_rainfall>=required_rainfall:
    return "Adequate Rainfall"
  else:
    return "Deficient Rainfall"
average_rainfall=float(input("Enter the average rainfall: "))
required_rainfall=float(input("Enter the required rainfall: "))
result=seasonal_rainfall_data(average_rainfall,required_rainfall)
print("Average Rainfall: ",average_rainfall)
print(result)
```

Problem 4: Social Media – Duplicate Account Detection

```python
def duplicate_acc(usernames):
    if len(usernames) != len(set(usernames)):
        return "Yes"
    else:
        return "No"

usernames = input("Enter all usernames: ").split()

result = duplicate_acc(usernames)
print("Duplicate Accounts Found:", result)
```

Problem 6: Agriculture – Premium Crop Price Filter

```python
def premium_crop_filter(prices):
    premium = [price for price in prices if price > 2000]
    return premium

prices = list(map(int, input("Enter crop prices: ").split()))

result = premium_crop_filter(prices)


print("Premium Crop Prices:", result)
```

Problem 7: System Monitoring – Application Health Checker

```python
def error_count(error):
  if error==0:
    return "Healthy"
  elif error<=5:
    return "Minor Issues"
  else:
    return "Critical Issues"
error=int(input("enter error count: "))
result=error_count(error)
print("Error Count: ",error)
print("System Status: ",result)
```

Problem 9: E-Learning – Student Attendance Eligibility System

```python
def calc_attendance(l,n):
  cp=l.count("P")
  return ((cp/n)*100)
def check_elgibility(perc):
  if perc>=75:
    return "Eligible"
  else:
    return "Not Eligible"
n = int(input("Enter number of days: "))
l = list(map(str.upper, input("Enter the attendance (P/A separated by space): ").split()))
perc=calc_attendance(l,n)
elgi=check_elgibility(perc)
print("Attendance Percentage: ",perc)
print("Exam Eligibility: ",elgi)
```

Problem 10: Smart Electricity Bill Analyzer

```python
def electricity_bill(units):
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"
    return bill, usage
units = int(input("Enter units consumed: "))
total_bill, usage_status = electricity_bill(units)
print("Total Bill: ₹", total_bill)
print("Usage Level:", usage_status)
```
